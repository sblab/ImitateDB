```python
df_hostInteractors = pd.read_csv(r'C:\Users\91956\OneDrive\Desktop\host interactors.csv', low_memory = False)
```

```python
df_pathogenDomains = df_pathogenDomains.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
df_interactions = df_interactions.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
df_hostInteractors = df_hostInteractors.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
df_interactorDomains = df_interactorDomains.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
df_interactions_pathogenDomains = df_interactions.merge(df_pathogenDomains,how = 'left',left_on = 'Pathogen ID',right_on = 'Pathogen ID')
```

```python
df_interactions_hostInteractors = df_interactions_pathogenDomains.merge(df_hostInteractors,how = 'left',left_on = 'Host ID',right_on = 'Entry')
```

```python
df_interactions_hostInteractors
```

```python
df_interactions_hostInteractors = df_interactions_hostInteractors.drop_duplicates()
df_interactions_hostInteractors = df_interactions_hostInteractors.dropna()
```

```python
df_interactions_hostInteractors
```

```python
df_interactions_interactors = df_interactions_hostInteractors.merge(df_interactorDomains,how = 'left',left_on = 'Interactor ID',right_on = 'Interactor ID')
```

```python
df_interactions_interactors
```

```python
df_interactions_interactors = df_interactions_interactors.drop_duplicates()
```

```python
df_interactions_interactors
```

```python
df_interactions_interactors = df_interactions_interactors.dropna()
```

```python
df_host_pathogen_domains = df_interactions_interactors[(df_interactions_interactors['PSSM-ID_x'].isnull() == False) & (df_interactions_interactors['PSSM-ID_y'].isnull() == False) ]
```
