```python
import Bio.ExPASy.ScanProsite as scanProsite
import json
import requests
import pandas as pd
import tqdm.notebook
```

```python
df_viral_pathogens = pd.read_excel(r'C:\Final data\VIRUS FINAL\viralpathogens.xlsx')
```

```python
PathogenList = list(df_viral_pathogens['Pathogen ID'].values)
```

```python
PathogenList = list(set(PathogenList))
```

```python
motifs_ = {}
```

```python
for i in tqdm.notebook.tqdm(PathogenList):
    try:
        print(i)
        data = scanProsite.scan(seq = i,mirror = "https://prosite.expasy.org",output = "json", skip='off')
        data_ = requests.get(data.url)
        data = data_.json()
        motifList = []
        for j in data['matchset']:
            motifList.append(j['signature_ac'])
        print(motifList)
        motifs[i] = motifList
    except:
        pass
```
